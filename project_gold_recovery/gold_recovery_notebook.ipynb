### Технологический процесс

Как золото получают из руды?  
Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.  
Опишем каждую стадию:  
1. _Флотация:_  
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
2. _Очистка:_  
Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### Расчёт эффективности

Вам нужно смоделировать процесс восстановления золота из золотосодержащей руды.  
Эффективность обогащения рассчитывается по формуле:  

$$Recovery = \frac {C\times (F - T)} {F\times (C - T)} \times 100\%$$

где:  
- $C$ — доля золота в концентрате после флотации/очистки;
- $F$ — доля золота в сырье/концентрате до флотации/очистки;
- $T$ — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.

### Метрика качества

Для решения задачи введём новую метрику качества — _sMAPE_ (англ. _Symmetric Mean Absolute Percentage Error_, «симметричное среднее абсолютное процентное отклонение»).
Она похожа на _MAE_, но выражается не в абсолютных величинах, а в относительных. Почему симметричная? Она одинаково учитывает масштаб и целевого признака, и предсказания.
Метрика _sMAPE_ вычисляется так:  

$$sMAPE = \frac 1 N \sum_{i=1}^{N} \frac {|y_i - \hat{y_i}|} {(|y_i| - \hat{|y_i|})\ /\ 2} \times 100\%$$

где:
- $y_i$ - значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество.  
- $\hat{y_i}$ значение предсказания для объекта с порядковым номером i, например, в тестовой выборке.  
- $N$ - количество объектов в выборке.  
- $\sum_{i=1}^{N}$ - суммирование по всем объектам выборки (i меняется от 1 до N). 

Нужно спрогнозировать сразу две величины:  
- эффективность обогащения чернового концентрата `rougher.output.recovery`;
- эффективность обогащения финального концентрата `final.output.recovery`.  

Итоговая метрика складывается из двух величин:  

$$Итоговое\ sMAPE = 25\% \times sMAPE(rougher) + 75\% \times sMAPE(final)$$

### Проверка корректности расчета метрики эффективности обогащения

```python
# функция расчета метрики эффективности по заданной формуле
def recovery(c_values, f_values, t_values):
#     print('c=', c_values, 'f=', f_values, 't=', t_values)
    return 100 * c_values * (f_values - t_values) / (f_values * (c_values - t_values))
```

```python
# функция с добавкой для предотвращения деления на 0
def recovery_corrected(c_values, f_values, t_values):
    return 100 * (c_values * (f_values - t_values) + 0.0001) / (f_values * (c_values - t_values) + 0.0001)
```

Однако определить формулы расчета для признаков не представляется возможным, хотя некоторая зависимость между направленным в технологический процесс объемом сернистого натрия и его содержанием в концентрате золота и поступающей руде существует. Очевидно, что сульфид натрия взаимодействует и с другими компонентами руды и это тоже должно учитываться в расчетной формуле.

```python
# функция отбора коррелированных признаков
def get_corr(data, targets_to_corr, feature_to_corr, treshold):
    correlated, coef = [], []
    for target in targets_to_corr:
        for feature in feature_to_corr:
            if target not in feature:
                # корреляция только двух признаков
                correlation = data[target].corr(data[feature])
                # учитявается и положительная и отрицательная зависимость
                if abs(correlation) > treshold:
                    correlated.append(feature)
                    coef.append(correlation)
    return correlated, coef
```

#### Решение проблемы выбросов

В данных имеются выбросы. Обрежем их использую 0.01-ый и 0.99-ый квантили на каждом признаке. Для этого напишем функцию. Чтобы не потерять данные, квантили будут определяться не по преобразуемой, а по исходной выборке.

```python
# функция обрезки выбросов
def cut_outliers(data_to_cut, columns, quantile_value):
    # копия датасета
    data_cutted = data_to_cut.copy()    
    for col in columns:
        # отсев строковых данных
        if data_cutted[col].dtypes == 'int64' or data_cutted[col].dtypes == 'float64':
            # расчет квантилей по исходной выборке
            quantiles = data_to_cut[col].quantile(q=[quantile_value, 1 - quantile_value])
            # обрезка датасета по квантилям
            data_cutted[col]  =\
            data_cutted.loc[(data_cutted[col] >= quantiles.iloc[0]) & (data_cutted[col] <= quantiles.iloc[1]), col].copy()
    # удаление пропусков
    data_cutted.dropna(axis=0, inplace=True)
    return data_cutted
```

#### Решение проблемы мультиколлинеарности

Среди объясняющих признаков имеется значительно число признаков имеющих сильную линейную зависимость между собой. Если такие признаки еще и сильно коррелированны с целевым признаком появляется мультиколлинеарность и прогноз моделей по таким признакам становиться неустойчивым.  
Оценим наличие проблемы мультиколлинеарности с помощью и коэффициента инфляции дисперсии _(VIF)_. Расчетная формула этого коэффициента для отдельного регрессора:

$$ VIF_i = \frac {1} {1 - R{^2}{_i}} $$
где:
- $R{^2}{_i}$ - коэффициент детерминации.  

Пороговым значением,  при котором мультиколлинеарность становится явно выражена, обычно принимается значение около 10.  
Возьмем часть признаков обучающей выборки (это признаки, имеющиеся в тестовой), а также один из целевых признаков `final.output.concentrate_au`. 

```python
# тестовая часть объясняющих признаков и один из целевых на данных обучающей выборки
features_to_vif = initial_features_names + ['final.output.concentrate_au']
features_to_vif.remove('date')
data_to_vif = StandardScaler().fit_transform(data_train_cutted[features_to_vif])
for i in range(len(features_to_vif)):
    # расчет vif для отдельных признаков
    vif = variance_inflation_factor(data_to_vif, i)
    # порог значимости равен 10
    if vif > 10:
        print(features_to_vif[i], vif)
```

Судя по тому как много признаков превысило принятое пороговое значение в данных присутствует мультиколлинеарность. Одним из способов борьбы является исключение ряда зависимых признаков из признакового пространства.  
Можно использовать следующий подход:
- 1 упорядочить признаки по степени корреляции с целевым;
- 2 взять самый коррелированный;
- 3 посчитать $R^2$ регрессии;
- 4 добавлять признаки по одному в соответствии со значимостью, пока $R^2$ не перестанет расти.  

Однако, у нас несколько целевых признаков и между объясняющими факторами может оказаться сложная взаимосвязь. Нам не хотелось бы терять признаки.  
Другим способом борьбы с мультиколлинеарностью является применение регуляризации. Но ее использование деформирует признаки, они теряют в интерпретируемости. А если учесть, что нашей целью в проекте является не столько получение точного прогноза по целевой метрике, сколько отлаживание технологического цикла, то использовать такие признаки в производстве окажется крайне затруднительным.  
В связи с этим основной мерой борьбы с мультиколлинеарностью стоит принять использование моделей, устойчивых к этому явлению и хорошо интерпретируемых.

### Сравнение распределения размеров гранул сырья на обучающей и тестовой выборках

Наиболее очевидным решением для сравнения двух выборок является проведение с помощью бутсрепа _A/B_-теста для сравнения типичных значений. Такой подход тем более актуален, что технологические процессы стремятся к унификации, что затрагивает и размеры гранул поступающего сырья.  
Для сравнения распределений размеров гранул сырья на обучающей и тестовой выборках выдвинем гипотезы:
- $H_0:$ Распределения размеров гранул сырья на обучающей и тестовой выборках не отличаются.
- $H_1:$ Распределения размеров гранул сырья на обучающей и тестовой выборках различны.

```python
samples_A = data_train['rougher.input.feed_size']
samples_B = data_test['rougher.input.feed_size']
# фактическая разность средних значений в выборках
AB_difference = samples_A.mean() - samples_B.mean()
print('Разница средних значений выборок:', AB_difference)
alpha = 0.05
state = np.random.RandomState(12345)

bootstrap_samples = 1000
count = 0
for i in range(bootstrap_samples):
    # взятие сэмпла из объедененной выборки
    unied_subsample = data_combined['rougher.input.feed_size'].sample(n=len(samples_A), replace=True, random_state=state)
    # имитация разбиения на тестовую и обучающую выборки
    subsample_A = unied_subsample.iloc[:len(samples_A)]
    subsample_B = unied_subsample.iloc[len(samples_A):]
    # подсчет превышений разницы средних
    bootstrap_difference = subsample_A.mean() - subsample_B.mean()
    if abs(bootstrap_difference) >= abs(AB_difference):
        count += 1

pvalue = 1. * count / bootstrap_samples
print('p-value =', pvalue)

if pvalue < alpha:
    print('Отвергаем нулевую гипотезу: скорее всего, выборки отличаются')
else:
    print('Не отвергаем нулевую гипотезу: скорее всего, выборки схожи.')
```

Итак, выборки отличаются. Однако, в задачах машинного обучения прогноз делается не для целых выборок, а для каждого объекта в отдельности, а значит нам достаточно, чтобы параметры его находились в пределах параметров объектов на которых модель обучалась. То есть, чтобы среднее тестовой выборки находилось в каких-то разумных границах, например, в пределах межквартильного размаха распределения признака в обучающей выборки. - Обучить разные модели и оценить их качество кросс-валидацией. Выбрать лучшую модель и проверить её на тестовой выборке.

### Написание функции для вычисления итоговой _sMAPE_

```python
# функция расчета smape
def smape(target_values, predicted_values):
    target, predict = target_values, predicted_values
    return (100. * (abs(target - predict) / ((abs(target) + abs(predict)) / 2)).sum() / len(target))
```

```python
# функция расчета общей smape
def smape_total(smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```

```python
# функция генерации признаков
def predict_feature(features, targets, models):
    # считывание названий целевых признаков
    try:
        targets_names = list(targets.columns)
    except:
        # атрибут targets может быть датафреймом или списком
        if type(targets) == list:
            targets_names = targets
        else:
            print('Названия целевых признаков отсутствуют.')
            return
    # удаление даты из выборки признаков
    try:
        features.drop('date', axis=1, inplace=True)
    except:
        pass
    # выборка со сгенерированными целевыми признаками
    features_extended = features.copy()    
    for col in targets_names:
        print(col)
        try:
            # проверка отсутствия и запись модели
            if models.get(col) is None:
                model = Pipeline([('scaler', StandardScaler()),
                                  ('pca', PCA(n_components=10, random_state=12345)),
                                  ('linear', LinearRegression())]).fit(features, targets[col])
                models[col] = {'model': model, 'learned_on': list(features.columns)}
                print('Модель записана.')
            try:
                # прогноз модели из словаря для нужного признака по записанному пространству признаков                
                features_extended[col] = models.get(col)['model'].predict(features[models.get(col)['learned_on']])
                print('Сделан прогноз.')
            except:
                print('Прогноз не возможен.') 
        except:
            print('Не передан словарь моделей.')
    return features_extended
```

#### Разработка алгоритма кросс-валидации

Качество обучения на кросс-валидации оценивается искусственной метрикой - _итоговой sMAPE_. Причем оценивается совокупный процесс, содержащий в себе несколько вложенных друг в друга процессов. Разобрать такую "матрешку" само по себе является нетривиальной задачей. А добавление сюда еще и кросс-валидации серьезно усложняет дело. Тем не менее, ниже был реализован алгоритм кросс-валидации с использованием _итоговой sMAPE_ для оценки прогнозов по отдельным сплитам.

```python
def learn_check_model(rougher_features_train,
                      rougher_feature_test,
                      final_feature_train,
                      final_feature_test,
                      target_train_unied,
                      target_test_unied,
                      base_model,
                      index_train_num=None,
                      index_test_num=None):
        # индексы временных выборок
        if index_train_num is None:            
            index_train = rougher_features_train.index
            index_test = rougher_feature_test.index
        else:
            try:
                # индексы для кросс-валидации
                index_train = rougher_features_train.index[index_train_num]
                index_test = rougher_feature_test.index[index_test_num]
            except:
                print('Индексы не переданы в функцию.')
                return        
      
        # Фаза 1: Создание выборок
        # выборки признаков этапа флотации
        rougher_features_train_folded = rougher_features_train.loc[index_train, :]
        rougher_features_test_folded = rougher_feature_test.loc[index_test, :]
        # выборки целевого признака rougher.output.concentrate_au этапа флотации
        rougher_target_concentrate_train_folded = target_train_unied.loc[index_train, 'rougher.output.concentrate_au']        
        rougher_target_concentrate_test_folded = target_test_unied.loc[index_test, 'rougher.output.concentrate_au']        
        # выборки целевого признака rougher.output.tail_au этапа флотации
        rougher_target_tail_train_folded = target_train_unied.loc[index_train, 'rougher.output.tail_au']
        rougher_target_tail_test_folded = target_test_unied.loc[index_test, 'rougher.output.tail_au']
        # выборка целевого признака rougher.output.recovery этапа флотации
        rougher_target_recovery = target_test_unied.loc[index_test, 'rougher.output.recovery']
        
        # выборки признаков этапа очистки
        final_features_train_folded = final_feature_train.loc[index_train, :]
        final_features_test_folded = final_feature_test.loc[index_test, :]
        # выборки целевого признака final.output.concentrate_au этапа очистки
        final_target_concentrate_train_folded = target_train_unied.loc[index_train, 'final.output.concentrate_au']
        final_target_concentrate_test_folded = target_test_unied.loc[index_test, 'final.output.concentrate_au']
        # выборки целевого признака final.output.tail_au этапа очистки
        final_target_tail_train_folded = target_train_unied.loc[index_train, 'final.output.tail_au']
        final_target_tail_test_folded = target_test_unied.loc[index_test, 'final.output.tail_au']
        # выборка целевого признака final.output.recovery этапа очистки
        final_target_recovery = target_test_unied.loc[index_test, 'final.output.recovery']
        
        # Фаза 2: Обучение моделей и прогноз по выборкам
        # обучение и предсказания по модели для признака rougher.output.concentrate_au
        rougher_concentrate_model = base_model
        rougher_concentrate_model.fit(rougher_features_train_folded, rougher_target_concentrate_train_folded)
        rougher_concentrate_predicted = rougher_concentrate_model.predict(rougher_features_test_folded)
        # обучение и предсказания по модели для признака rougher.output.tail_au
        rougher_tail_model = base_model
        rougher_tail_model.fit(rougher_features_train_folded, rougher_target_tail_train_folded)
        rougher_tail_predicted = rougher_tail_model.predict(rougher_features_test_folded)
        # обучение и предсказания по модели для признака final.output.concentrate_au
        final_concentrate_model = base_model
        final_concentrate_model.fit(final_features_train_folded, final_target_concentrate_train_folded)
        final_concentrate_predicted = final_concentrate_model.predict(final_features_test_folded)
        # обучение и предсказания по модели для признака final.output.tail_au
        final_tail_model = base_model
        final_tail_model.fit(final_features_train_folded, final_target_tail_train_folded)
        final_tail_predicted = final_tail_model.predict(final_features_test_folded)
        
        # Фаза 3: Расчет коэффициента эффективности по выборкам кросс-валидации
        rougher_recovery_predicted = recovery(
            rougher_concentrate_predicted,
            target_test_unied.loc[index_test, 'rougher.input.feed_au'],
            rougher_tail_predicted
            )             
        final_recovery_predicted = recovery(
            final_concentrate_predicted,            
#             rougher_concentrate_predicted,
            # замена значений по сырью этапа очистки на начальные
            target_test_unied.loc[index_test, 'rougher.input.feed_au'],
            final_tail_predicted
        )
        # Фаза 4: Расчет погрешности коэффициента эффективности по выборкам
        rougher_recovery_smape = smape(rougher_target_recovery, rougher_recovery_predicted) 
        final_recovery_smape = smape(final_target_recovery, final_recovery_predicted) 
        recovery_smape_total = smape_total(rougher_recovery_smape, final_recovery_smape)
#         print(rougher_recovery_smape, final_recovery_smape, recovery_smape_total)
        
        return base_model, recovery_smape_total
```

```python
state = np.random.RandomState(12345)

cv_models = {
    'Lasso': Pipeline([('scaler', StandardScaler()), ('lasso', Lasso(random_state=state))]),
    'Ridge': Pipeline([('scaler', StandardScaler()), ('ridge', Ridge(random_state=state))]),
    'KNeighbors': Pipeline([('scaler', StandardScaler()), ('pca', PCA(n_components=10, random_state=state)), ('knn', KNeighborsRegressor())]),
    'ElasticNet': Pipeline([('scaler', StandardScaler()), ('elastik', ElasticNet(random_state=state))]),
    'AdaBoost': Pipeline([('scaler', StandardScaler()), ('pca', PCA(n_components=10, random_state=state)), ('ada', AdaBoostRegressor(n_estimators=300, random_state=state))]),
    'RandomForest': Pipeline([('scaler', StandardScaler()), ('pca', PCA(n_components=10, random_state=state)), ('rf', RandomForestRegressor(n_estimators=300, max_depth=8, random_state=state))]),
    'LinearPCA': Pipeline([('scaler', StandardScaler()), ('pca', PCA(n_components=5, random_state=state)),  ('linear', LinearRegression())]),
    'LinearRegression': Pipeline([('scaler', StandardScaler()), ('linear', LinearRegression())]),
    'DummyRegressor': Pipeline([('scaler', StandardScaler()), ('dummy', DummyRegressor())])
}

# списки под результаты
titles, scores = [], []
folds = 5
for title, model_piped in cv_models.items():    
    model_to_cv = model_piped
    titles.append(title)
    # разбиение данных на фолды
    k_fold = KFold(n_splits=folds, shuffle=True, random_state=state)
    
    cv_score = []
    # снятие порядковых номеров с индексов и передеча в сплиты
    for train_num, valid_num in k_fold.split(rougher_features_train_scaled):        
        # функция для обучения
        model_learned, score = learn_check_model(rougher_features_train_scaled,
                                                 rougher_features_train_scaled,
                                                 final_feature_train_scaled,
                                                 final_feature_train_scaled,
                                                 target_train,
                                                 target_train,
                                                 model_to_cv,
                                                 train_num,
                                                 valid_num)
        cv_score.append(score)
    # запись среднего smape на кросс-валидации
    scores.append(np.mean(cv_score))
# таблица показателей моделей на кросс-валидации
cv_results = pd.DataFrame({'Model title': titles, 'sMAPE': scores})
cv_results
```

### Проверка лучшей модели на тестовой выборке

## Подготовка данных

### Удаление неинформативных данных

```python
# удаление столбцов с идентификаторами и фамилиями
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

### Стилистические исправления

```python
# функция преобразования стиля записи переменных
def to_snake_case(stroke, delimiter='_'):
    stroke = stroke[0].lower() + stroke[1:]
    for letter in string.ascii_uppercase:
        stroke = stroke.replace(letter, delimiter + letter.lower())
    return stroke
```

```python
# исправление имен столбцов
new_columns = []
for col in df.columns:
    new_columns.append(to_snake_case(col))
df.columns = new_columns
```

### Создание выборок

```python
# выделение предсказательных и целевого признаков
features = df.drop(['exited'], axis=1)
target = df['exited']

# выделение обучающей и тестовой выборкок
features_train, features_test, target_train, target_test =\
train_test_split(features, target, test_size=0.2, stratify=target, random_state=12345)
```

Так как мы планируем задействовать поиск по сетке, то валидационная выборка будет использоваться только для проверки отдельных метрик в демонстрационных целях и не будет исключена из обучения.

```python
# выделение валидационной выборки для проверки метрик
features_demo_train, features_valid, target_demo_train, target_valid =\
train_test_split(features_train.copy(), target_train.copy(), test_size=0.2, stratify=target_train.copy(), random_state=12345)
```

### Кодирование и масштабирование признаков

Кодируем категориальные признаки и несколько признаков с дискретными числовыми значениями. Так как в тестовых данных могут встретиться неизвестные классы, оставляем все столбцы закодированного признака.

```python
# функция кодирования обучающей и тестовой выборок
def train_test_one_hot(train, test, categorial_names):
    for categorial in categorial_names:
        # категории в обучающей и тестовой выборке не отличаются
        if set(test[categorial]) == set(train[categorial]):
            train_coded = pd.get_dummies(train[categorial], prefix=categorial, prefix_sep='_', drop_first=True)
            test_coded = pd.get_dummies(test[categorial], prefix=categorial, prefix_sep='_', drop_first=True)
        else:        
            # обучающая выборка кодируется с запасным классом
            train_coded = pd.get_dummies(train[categorial], prefix=categorial, prefix_sep='_', dummy_na=True)
            # удаление артефактов в названиях столбцов
            train_coded.columns = train_coded.columns.str.replace(r'\.0', r'', regex=True)          
            test_coded = pd.get_dummies(test[categorial], prefix=categorial, prefix_sep='_')            
            # названия новых признаков в тестовой выборке
            na_classes_names = list(set(test_coded.columns) - set(train_coded.columns))
            # пустой массив с индексом от тестовой выборки
            test_temp = pd.DataFrame(index=test_coded.index)           
            for column in train_coded.columns:
                try:                    
                    test_temp[column] = test_coded[column]    
                # вставляем в тестовую выборку отсутствующие столбцы
                except:
                    test_temp[column] = 0

            # удаление лишнего признака в первом столбце    
            train_coded = train_coded.iloc[:, 1:]
            test_temp = test_temp.iloc[:, 1:]

            # новые категории в тестовой выборке
            na_classes = list(set(test[categorial]) - set(train[categorial]))
            # проверка и запись новых категорий 
            if len(na_classes) > 0:
                test_temp.iloc[:, -1] = (test_coded.loc[:, na_classes_names].sum(axis=1) > 0).astype('int64')
            # перезапись тестовой выборки
            test_coded = test_temp
        
        # объединение с обучающей выборкой
        train = train.join(train_coded)
        train = train.drop([categorial], axis=1)  
        # объединение с тестовой выборкой        
        test = test.join(test_coded)
        test = test.drop([categorial], axis=1)
        # смена регистра в названиях столбцов
        train.rename(str.lower, axis='columns', inplace=True)
        test.rename(str.lower, axis='columns', inplace=True)
    return train, test
```

```python
# признаки для one-hot-кодирования
categorials = ['geography', 'gender', 'tenure', 'num_of_products']

# one-hot-кодирование обучающей и тестовой выборок
features_train, features_test =\
train_test_one_hot(features_train, features_test, categorial_names=categorials)

# one-hot-кодирование валидационной выборки
features_demo_train, features_valid =\
train_test_one_hot(features_demo_train, features_valid, categorial_names=categorials)
```

Масштабируем числовые признаки.

```python
# масштабирование числовых признаков
numeric = ['credit_score', 'age', 'balance', 'estimated_salary']

# настройка масштабирование на обучающей выборке
scaler = StandardScaler()
scaler.fit(features_train[numeric])
features_train[numeric] = scaler.transform(features_train[numeric])
# масштабирование тестовой выборки
features_test[numeric] = scaler.transform(features_test[numeric])

# настройка масштабирование на части обучающей выборке
scaler_demo = StandardScaler()
scaler_demo.fit(features_demo_train[numeric])
features_demo_train[numeric] = scaler_demo.transform(features_demo_train[numeric])
# масштабирование валидационной выборки
features_valid[numeric] = scaler_demo.transform(features_valid[numeric])

# подавление предупреждений
#pd.options.mode.chained_assignment = None
features_train.head()
```

### Увеличение выборки

Для устранения дисбаланса классов используем *upsampling* и увеличим размер меньшего класса до размера большего.

```python
# функция апсэмплинга
def upsample(features, target):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    # дисбаланс классов
    repeat = round(len(features_zeros) / len(features_ones))
    # увеличение меньшего класса
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    # смешивание объектов
    return shuffle(features_upsampled, target_upsampled, random_state=12345)

# расширенный датасет
features_sampled, target_sampled = upsample(features_train, target_train)
```

```python
# поиск по сетке с одним гиперпараметром для случайного леса
rf_gs_ups = GridSearchCV(RandomForestClassifier(n_estimators=30, random_state=12345),
                     param_grid={'max_depth': [depth for depth in range(1, 20)]},
                     scoring='f1'
                    )
rf_gs_ups.fit(features_sampled, target_sampled)
# данные лога обучения rf_gs_ups.cv_results_
print('Параметры лучшей модели:\n', rf_gs_ups.best_estimator_)
# значения метрик лучшей модели на валидации
print('\nЗначение F1-меры лучшей модели на валидации:', rf_gs_ups.best_score_)
```

### ROC-кривая

Использование *ROC*-кривой также можно рассматривать как косвенный метод снижения дисбаланса. И хотя впрямую соотношения классов не учитывается, но обучение только на одном целевом классе снимает эту проблему. С помощью функции `GridSearchCV` можно провести обучение модели с использованием нескольких метрик - *ROC AUC* и *F1*-меры, причем последняя останется нашей целевой метрикой. Усилим модель введением параметра `class_weight='balanced'`.

```python
# поиск по сетке с одним гиперпараметром для случайного леса
rf_gs_rc = GridSearchCV(RandomForestClassifier(class_weight='balanced', n_estimators=30, random_state=12345),
                     param_grid={'max_depth': [depth for depth in range(1, 20)]},
                     scoring={'ROC_AUC_score': 'roc_auc', 'F1_score': 'f1'},
                     refit='F1_score'
                    )
rf_gs_rc.fit(features_train, target_train)
# данные лога обучения rf_gs.cv_results_
print('Параметры лучшей модели:\n', rf_gs_rc.best_estimator_)
# номер лучшей модели
num_best_model = rf_gs_rc.cv_results_['rank_test_F1_score'][0]
# значения метрик лучшей модели на валидации
print('\nЗначение F1-меры лучшей модели на валидации:', rf_gs_rc.best_score_)
print('Значение ROC AUC лучшей модели на валидации:', rf_gs_rc.cv_results_['mean_test_ROC_AUC_score'][num_best_model - 1])
```

```python
# график изменение значений метрик при обучении
fig = go.Figure()
# данные лога обучения rf_gs_rc.cv_results_
fig.add_trace(go.Scatter(x=rf_gs_rc.cv_results_['param_max_depth'].data,
                         y=rf_gs_rc.cv_results_['mean_test_ROC_AUC_score'],
                         name='ROC AUC'
                        )
             )
fig.add_trace(go.Scatter(x=rf_gs_rc.cv_results_['param_max_depth'].data,
                         y=rf_gs_rc.cv_results_['mean_test_F1_score'],
                         name=u'F1-мера'
                        )
             )
fig.update_layout(title_text=u'Изменение значений метрик ROC AUC и F1-меры <br>в процессе обучения модели случайного леса')
fig.update_xaxes(title_text=u'Максимальная глубина дерева (параметр max_depth), ед.')
fig.update_yaxes(title_text=u'Метрика, ед.')
fig.show(renderer='png')
```
